```python
import numpy as np
from scipy.optimize import root
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```python
a = 28 # α 
b = 7 # β
g = 3 # γ
t = 1 # θ
m = 1 # μ
v = 3 # ν
```

```python
# 1
# Решение
def equation(x, m, a):
    return (m + x)**3 - 2*x**2 - a*x - 17

solution = root(equation, x0=[0], args=(m, a))
x = solution.x[0]

print("Корень x =", x)
```

```python
# 2
# Решение
def polynomial(x, a, g, t, m, v):
    return a*x**3 - (g + t)*x**2 - m*x + v

def equation(x):
    return polynomial(x, a, g, t, m, v)

x_values = np.linspace(-10, 10, 100)
y_values = polynomial(x_values, a, g, t, m, v)

roots = fsolve(equation, x0=[-5, 5])

print("Корни x =", roots)

plt.plot(x_values, y_values)
plt.axhline(0, color='black', linestyle='--')
plt.scatter(roots, np.zeros_like(roots), color='red')
plt.xlabel('x')
plt.ylabel('Корни')
plt.title('Графическое решение')
plt.grid(True)
plt.show()
```

```python
# 3
# Решение
def equations(x):
    x_1, x_2, x_3, x_4 = x[0], x[1], x[2], x[3]
    eq1 = 3 * x_1 - x_2 + x_3 + g * x_4 - a
    eq2 = x_1 - m * x_2 + v * x_3 + 4 * x_4 - b
    eq3 = -5 * x_1 - x_3 -7 * x_4 + a
    eq4 = x_1 - 6 * x_2 + a * x_3 + 6 * x_4 - 3
    return [eq1, eq2, eq3, eq4]

solution = fsolve(equations, [0, 0, 0, 0])

solution
```
