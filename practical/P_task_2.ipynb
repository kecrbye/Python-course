```python
# Задание 1.14
x = ['2016-10-03', '2016-10-04', '2016-10-05', '2016-10-06', '2016-10-07']
y = [772.5, 776.4, 776.5, 776.9, 775.1]

# График с нужными точками
plt.plot(x, y, color="red", marker = 'o')

# Настройки сетки
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth= '0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth= '0.5', color='black')

# Настройки подписей к осям и графику
plt.title('Closing stock value of Alphabet Inc.')
plt.xlabel('Date')
plt.ylabel('Closing value')

# Ограничения для осей
plt.xlim('2016-10-03', '2016-10-07')
plt.ylim(772.5, 777.0)
```

```python
# Задание 2.14
# Данные подобрал для более наглядной визуализации, как на графике, где значения для женщин примерно в 2 раза больше
y_male = [24, 32, 28, 29, 25]
y_female = [48, 60, 64, 57, 52]
std_male = [4, 6, 9, 6, 4]
std_female = [3, 5, 5, 7, 2]
groups = [f"Group{i}" for i in range(1, 6)]

# Настройки столбчатых диаграмм
plt.bar(groups, y_female, yerr = std_female, width = 0.4, color = 'green', ecolor = 'g', capsize = 4)
plt.bar(groups, y_male, yerr = std_male, width = 0.4, color = 'red', ecolor = 'b', capsize = 4)

# Настройки подписей и легенды
plt.title('Scores by group \n and gender')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.legend(['Women', 'Men'])

# Ограничение по вертикальной оси
plt.ylim(0, 80)

plt.show()
```

```python
# 1. Построить несколько графиков на одном поле

fig ,axes = plt.subplots(nrows = 1, ncols = 2, figsize=(8,6))

# Распределение рейтинга и потенциала игроков
sns.histplot(df_player_att['overall_rating'], ax = axes[0], color = 'red')
sns.histplot(df_player_att['potential'], ax = axes[1], color = 'orange')

plt.ylabel('')
```

```python
# 2. Построить диаграмму для категориальных данных 
# Распределение рейтинга атаки и защиты игроков

fig ,axes = plt.subplots(nrows = 1, ncols = 2, figsize=(9,5))

attack_work_rate = df_player_att[df_player_att['attacking_work_rate'].isin(['medium','high','low'])]
defence_work_rate = df_player_att[df_player_att['defensive_work_rate'].isin(['medium','high','low'])]

sns.histplot(attack_work_rate['attacking_work_rate'], ax = axes[0], color = 'red')
sns.histplot(defence_work_rate['defensive_work_rate'], ax = axes[1], color = 'green')

plt.ylabel('')
```

```python
# 3. Построить столбчатые и круговые диаграммы

# Распределение матчей по лигам в %  

merged_match = pd.merge(df_match, df_league, left_on = 'league_id', right_on = 'id', how = 'inner')
merged_match = merged_match.rename(columns = {'name' : 'ligue'})
merged_match['ligue'].value_counts(normalize = True).plot(kind = 'pie', autopct = '%1.1f%%', ylabel = '')
```

```python
# Сколько лиг представлено от каждой страны: для каждой страны одна уникальная лига

merged_df = pd.merge(df_country, df_league, left_on='id', right_on='country_id', how='inner')
merged_df = merged_df.rename(columns = {'name_x' : 'country', 'name_y' : 'ligue'})

merged_df.groupby('country').agg({'ligue' : 'count'}).plot(kind = 'bar')
```

```python
# 4. Реализовать цветовую сетку данных

players = df_player_att.select_dtypes(include=['float64', 'int64']) \
    .drop(columns=['id', 'player_fifa_api_id', 'player_api_id'])
players_corr = players.corr()

# Тепловая карта матрицы корреляции характеристик игроков
sns.heatmap(players_corr)
```

```python
# 5. Построить 3D-график

# 3D-график для характеристик футбольных вратарей
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(players['gk_diving'], players['gk_kicking'], players['gk_reflexes'])

plt.show()
```
